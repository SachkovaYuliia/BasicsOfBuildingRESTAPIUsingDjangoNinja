```python
# 1. Вивести інформацію про набір даних, основні статистичні характеристики, типи ознакfrom google.colab import files

from google.colab import files
import pandas as pd

# Завантажуємо файл
uploaded = files.upload()

# Зчитуємо файл у DataFrame
file_name = list(uploaded.keys())[0]
df = pd.read_csv(file_name)

# Перевіряємо завантажені дані
print(df.head())

# Виводимо інформацію про набір даних
print("\nІнформація про набір даних:")
df.info()

# Основні статистичні характеристики
print("\nОсновні статистичні характеристики:")
print(df.describe())

print("\nКількість унікальних значень у кожному стовпці:")
print(df.nunique())
```

```python
# Визначення категоріальних та кількісних ознак
categorical_features = df.select_dtypes(include=['object']).columns
numerical_features = df.select_dtypes(include=['number']).columns
print("Категоріальні ознаки:", categorical_features)
print("Кількісні ознаки:", numerical_features)
```

```python
# а) Мінімальна ціна авто кожного року реєстрації

min_price = df.groupby('year')['price'].min()
print(min_price)
```

```python
# б) Авто не старіші 5 років, напівавтоматична коробка передач, пробіг > 20 тис миль

from datetime import datetime

year_threshold = datetime.now().year - 5
filtered_cars = df[(df['year'] >= year_threshold) & (df['transmission'] == 'Semi-Auto') & (df['mileage'] > 20000)]
print(filtered_cars)
```

```python
# в) Додавання стовпця із загальною витратою пального (загальний пробіг автомобіля (у милях) ділимо на кількіст миль, яку проїжджає авто на 1 галоні пального)

df["total_fuel_consumed"] = df["mileage"] / df["mpg"]
print(df.head())
```

```python
# г) Додавання стовпця із середнім об'ємом двигуна для моделі
df["avg_mpg_per_model"] = df.groupby("model")["mpg"].transform("mean")
print(df.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

uploaded = files.upload()

file_name = list(uploaded.keys())[0]
df = pd.read_csv(file_name)

print(df.head())
```

```python
# а) Стовпчикова діаграма кількості пінгвінів кожного виду
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x="species")
plt.title("Кількість пінгвінів кожного виду")
plt.show()
```

```python
# б) Мінімальна довжина дзьоба пінгвінів кожного виду
min_culmen_length = df.groupby("species")["culmen_length_mm"].min()
min_culmen_length.plot(kind="bar", title="Мінімальна довжина дзьоба")
plt.show()
```

```python
# в) Середня вага пінгвінів кожного виду за статтю
df.dropna(subset=["sex", "body_mass_g"], inplace=True)
plt.figure(figsize=(8, 5))
sns.barplot(data=df, x="species", y="body_mass_g", hue="sex", estimator=np.mean)
plt.title("Середня вага пінгвінів за видами та статтю")
plt.show()
```

```python
# 2. Гістограма глибини дзьоба
plt.figure(figsize=(8, 5))
sns.histplot(data=df, x="culmen_depth_mm", hue="species", kde=True)
plt.title("Гістограма глибини дзьоба")
plt.show()
```

```python
# 3. Діаграма розмаху довжини ласт
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x="species", y="flipper_length_mm")
plt.title("Діаграма розмаху довжини ласт")
plt.show()
```

```python
# 4. Діаграми розсіювання
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x="culmen_length_mm", y="culmen_depth_mm")
plt.title("Залежність довжини і глибини дзьоба")
plt.show()

plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x="flipper_length_mm", y="body_mass_g")
plt.title("Залежність ваги і довжини ласт")
plt.show()
```

```python
# Обчислення коефіцієнтів кореляції
correlation_culmen = df[["culmen_length_mm", "culmen_depth_mm"]].corr()
correlation_flipper_body = df[["flipper_length_mm", "body_mass_g"]].corr()
print("Кореляція між довжиною і глибиною дзьоба:", correlation_culmen)
print("Кореляція між довжиною ласт і вагою:", correlation_flipper_body)
```
