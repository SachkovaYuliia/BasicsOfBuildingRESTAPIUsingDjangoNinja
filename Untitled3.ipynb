> а)Файл StudentsPerformance.csv
1. Вивести інформацію про набір даних, основні статистичні характеристики, типи ознак. Які ознаки є категоріальними, а які – кількісними?



```python
from google.colab import files
import pandas as pd

from google.colab import files
import pandas as pd

# Завантажуємо файл
uploaded = files.upload()

# Зчитуємо файл у DataFrame
file_name = list(uploaded.keys())[0]
df = pd.read_csv(file_name)

# Перевіряємо завантажені дані
print(df.head())

# Виводимо інформацію про набір даних
print("\nІнформація про набір даних:")
df.info()

# Основні статистичні характеристики
print("\nОсновні статистичні характеристики:")
print(df.describe())

print("\nКількість унікальних значень у кожному стовпці:")
print(df.nunique())
```

```python
# Визначаємо категоріальні та кількісні ознаки
categorical_features = df.select_dtypes(include=['object']).columns
numerical_features = df.select_dtypes(include=['number']).columns

print("\nКатегоріальні ознаки:", list(categorical_features))
print("Кількісні ознаки:", list(numerical_features))
```

2. Використовуючи початковий DataFrame:

```python
# 2.a. Знаходимо максимальний бал з математики в залежності від рівня освіти батьків
max_math_by_education = df.groupby('parental level of education')['math score'].max()
print("\nМаксимальний бал з математики залежно від рівня освіти батьків:")
print(max_math_by_education)
```

```python
# 2.b. Рахуємо кількість студентів-чоловіків, що пройшли підготовчі курси і отримали за читання не менше 75 балів
male_students = df[(df['gender'] == 'male') &
                   (df['test preparation course'] == 'completed') &
                   (df['reading score'] >= 75)]
male_students_count = len(male_students)
print("\nКількість студентів-чоловіків, що пройшли підготовчі курси і отримали за читання не менше 75 балів:", male_students_count)
```

```python
# 2.c. Додаємо новий стовпець, який містить середній бал студента
df['average score'] = df[['math score', 'reading score', 'writing score']].mean(axis=1)
print("\nДодано новий стовпець 'average score':")
print(df.head())
```

```python
# 2.d. Додаємо новий стовпець, який містить кількість студентів даної расової/етнічної групи
df['group count'] = df.groupby('race/ethnicity')['race/ethnicity'].transform('count')
print("\nДодано новий стовпець 'group count':")
print(df.head())
```

б) Файл telecom.csv.
1. Побудувати стовпчикові діаграми, на яких відобразити:


```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

uploaded = files.upload()

file_name = list(uploaded.keys())[0]
df = pd.read_csv(file_name)

print(df.head())
```

```python
# a) Кількість клієнтів з підключеним роумінгом та без
import matplotlib.pyplot as plt

roaming_counts = df['international plan'].value_counts()
roaming_counts.plot(kind='bar', color=['blue', 'yellow'])
plt.title('Кількість клієнтів з роумінгом і без')
plt.xlabel('Роумінг (так/ні)')
plt.ylabel('Кількість клієнтів')
plt.xticks(rotation=0)
plt.show()
```

```python
# b) Максимальна кількість хвилин у нічний час для клієнтів з роумінгом і без
max_night_minutes = df.groupby('international plan')['total night minutes'].max()
max_night_minutes.plot(kind='bar', color=['blue', 'yellow'])
plt.title('Максимальна кількість хвилин у нічний час (роумінг)')
plt.xlabel('Роумінг (так/ні)')
plt.ylabel('Максимальні хвилини')
plt.xticks(rotation=0)
plt.show()
```

```python
# c) Середня кількість міжнародних дзвінків для клієнтів з роумінгом та без, з урахуванням лояльності
# Додаємо стовпець лояльності
df['loyalty'] = (df['customer service calls'] <= 3) & (df['churn'] == 0)

# Перевіряємо результат
print(df[['customer service calls', 'churn', 'loyalty']].head())

# Середня кількість міжнародних дзвінків для клієнтів з роумінгом та без, з урахуванням лояльності
avg_int_calls = df.groupby(['international plan', 'loyalty'])['total intl calls'].mean().unstack()
avg_int_calls.plot(kind='bar', stacked=True, color=['blue', 'yellow'])
plt.title('Середня кількість міжнародних дзвінків')
plt.xlabel('Роумінг (так/ні)')
plt.ylabel('Середня кількість дзвінків')
plt.xticks(rotation=0)
plt.legend(title='Лояльність')
plt.show()
```

3. Діаграма хвилин на міжнародні дзвінки

```python
# Загальна діаграма
plt.boxplot(df['total intl minutes'], vert=False, patch_artist=True, boxprops=dict(facecolor='blue'))
plt.title('Хвилини на міжнародні дзвінки (загальна)')
plt.xlabel('Хвилини')
plt.show()
```

```python
# В залежності від роумінгу
sns.boxplot(data=df, x='total intl minutes', y='international plan', palette='Set2')
plt.title('Хвилини на міжнародні дзвінки (за роумінгом)')
plt.xlabel('Хвилини')
plt.ylabel('Роумінг')
plt.show()
```

```python
# Визначення викидів
q1 = df['total intl minutes'].quantile(0.25)
q3 = df['total intl minutes'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr
outliers = df[(df['total intl minutes'] < lower_bound) | (df['total intl minutes'] > upper_bound)]
print("Кількість викидів:", len(outliers))
```

4. Діаграми розсіювання

```python
# a) Хвилини і дзвінки в нічний час
sns.scatterplot(data=df, x='total night minutes', y='total night calls')
plt.title('Залежність: Хвилини і дзвінки в нічний час')
plt.xlabel('Нічні хвилини')
plt.ylabel('Нічні дзвінки')
plt.show()
```

```python
# b) Хвилини і оплата в нічний час
sns.scatterplot(data=df, x='total night minutes', y='total night charge')
plt.title('Залежність: Хвилини і оплата в нічний час')
plt.xlabel('Нічні хвилини')
plt.ylabel('Нічна оплата')
plt.show()
```

```python
# Коефіцієнт кореляції
corr_night_minutes_calls = df['total night minutes'].corr(df['total night calls'])
corr_night_minutes_charge = df['total night minutes'].corr(df['total night charge'])

print("Коефіцієнт кореляції між нічними хвилинами і нічними дзвінками:", corr_night_minutes_calls)
print("Коефіцієнт кореляції між нічними хвилинами і нічною оплатою:", corr_night_minutes_charge)
```
